```python
import pickle
```

```python
with open("MVImgNet_dataset_val_folder_st.pkl", "rb") as f:
    data = pickle.load(f)
```

```python
data
```

```python
print(len(data))
```

```python
num_example_per_shard = len(data) // 16

shards = []
for i in range(16):
    if i == 15:
        shards.append(data[i*num_example_per_shard:])
    else:
        shards.append(data[i*num_example_per_shard:(i+1)*num_example_per_shard])
```

```python
print(len(shards))
```

```python
import os

for i, shard in enumerate(shards):
    shard_text = []
    for sample in shard:
        filepath = os.path.join("mvimagenet_txt", sample["class_label"], sample["instance_id"], "text", "caption.txt")
        with open(filepath, "r") as f:
            caption = f.read()
            caption += "|" + sample["class_label"]
            caption += "|" + sample["instance_id"]
            shard_text.append(caption)

    with open("mv_shard_{}.txt".format(i), "w") as f:
        for caption in shard_text:
            f.write(caption + "\n")
```
